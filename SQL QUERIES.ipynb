### SQL QUERIES

Q.Retrieve the number of different colors available for each category

```python
%%sql
SELECT category,COUNT(DISTINCT color) as n_color
FROM shoes_project
GROUP BY category;
```

Q. Retrieve all shoe names and their corresponding prices for men's shoes.

```python
%%sql
SELECT name,price
FROM shoes_project
WHERE category LIKE 'men%'
LIMIT 10
```

Q. Write a query to find the most expensive men's shoe

```python
%%sql
SELECT name,price
FROM shoes_project
WHERE category LIKE 'men%'
ORDER by price DESC
LIMIT 10
```

 Q.Write a query that retrieves the count of sizes for all styles.

```python
%%sql
SELECT sc,COUNT(DISTINCT size)as size_count
FROM shoes_project
GROUP BY sc
```

Q.Write a query to list all styles with their associated colors.

```python
%%sql
SELECT sc ,GROUP_CONCAT(color) AS associated_color
FROM shoes_project
GROUP BY sc;
```

Q. Write a to find styles that have more than one color.

```python
%%sql
SELECT sc 
FROM shoes_project
GROUP BY sc
HAVING COUNT('black')>1;
```

Q. Write a query to find the count of sizes available for each color for a specific style code.

```python
%%sql
SELECT sc,color,COUNT('UK41') AS size_count
FROM shoes_project
WHERE sc='Style: FJ4170-002'
GROUP BY sc,color
```

Q. Write a query that counts the number of reviews for each product.

```python
%%sql
SELECT sc,name,COUNT(*) AS review_count
FROM shoes_project
GROUP BY sc,name;
```

Q.Write a query that  calculates the average comfort rating for each size

```python
%%sql
SELECT size,rating AS average_comfort_rating
FROM shoes_project
GROUP BY size;
```

Q. Retrieve products with high star ratings (e.g., 4 stars or above)



```python
%%sql
SELECT name
FROM shoes_project
GROUP BY name
HAVING rating>=4
```
